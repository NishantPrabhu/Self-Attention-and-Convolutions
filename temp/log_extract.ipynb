```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
learned_2d = os.listdir("../main/cifar10/learned_2d")
gaussian = os.listdir('../main/cifar10/gaussian')
resnet = os.listdir('../main/cifar10/resnet')

folders = learned_2d + gaussian + resnet
keys = ['l2d_'+name for name in learned_2d] + ['g_'+name for name in gaussian] + ['resnet_'+name for name in resnet]
```

```python
data = {}

for f in learned_2d:
    file = os.path.join('../main/cifar10/learned_2d', f, 'trainlogs.txt')
    k = 'l2d_'+f
    data[k] = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    with open(file, 'r') as t:
        lines = t.read().split('\n')
    for l in lines:
        if l.startswith('[TRAIN] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['train_loss'].append(float(loss))
            data[k]['train_acc'].append(float(acc))
        elif l.startswith('[VAL] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['val_loss'].append(float(loss))
            data[k]['val_acc'].append(float(acc))
            
for f in gaussian:
    file = os.path.join('../main/cifar10/gaussian', f, 'trainlogs.txt')
    k = 'g_'+f
    data[k] = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    with open(file, 'r') as t:
        lines = t.read().split('\n')
    for l in lines:
        if l.startswith('[TRAIN] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['train_loss'].append(float(loss))
            data[k]['train_acc'].append(float(acc))
        elif l.startswith('[VAL] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['val_loss'].append(float(loss))
            data[k]['val_acc'].append(float(acc))
            
for f in resnet:
    file = os.path.join('../main/cifar10/resnet', f, 'trainlogs.txt')
    k = 'resnet_'+f
    data[k] = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    with open(file, 'r') as t:
        lines = t.read().split('\n')
    for l in lines:
        if l.startswith('[TRAIN] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['train_loss'].append(float(loss))
            data[k]['train_acc'].append(float(acc))
        elif l.startswith('[VAL] [Loss]'):
            loss, acc = re.findall('\d{1}\.\d{4}', l)
            data[k]['val_loss'].append(float(loss))
            data[k]['val_acc'].append(float(acc))
```

```python
# 9 heads plot
resnet = data['resnet_2021-01-11_12-49_resnet18_9463']['val_acc']
l2d_nqc = data['l2d_2021-01-01_14-17_normal_nqc_9036']['val_acc']
l2d_qc = data['l2d_2020-12-31_16-43_normal_qc_8636']['val_acc']
gaus = data['g_2020-12-31_22-09_normal_9365']['val_acc'][:-1]
x = np.arange(1, 311, 10)

plt.figure(figsize=(8, 6))
plt.plot(x, resnet, color='g', linewidth=1, marker='.', label='ResNet18')
plt.plot(x, gaus, color='orange', linewidth=1, marker='.', label='Quadratic embedding')
plt.plot(x, l2d_nqc, color='b', linewidth=1, marker='.', label='Learned embedding w/o content')
plt.plot(x, l2d_qc, color='r', alpha=0.8, linewidth=1, marker='.', label='Learned embedding w/ content')
plt.legend()
plt.xlabel('Epoch', fontsize=14, labelpad=10)
plt.ylabel('Test accuracy', fontsize=14, labelpad=10)
plt.grid(alpha=0.4)
plt.savefig('../main/results/paper_models_9h.pdf', pad_inches=0.05)
plt.show()
```
